```python
import pandas as pd

# Load 
train = pd.read_csv('C:/Users/Riyad/projects/fake_news/train_cleaned.csv')
val = pd.read_csv('C:/Users/Riyad/projects/fake_news/val_cleaned.csv')
test = pd.read_csv('C:/Users/Riyad/projects/fake_news/test_cleaned.csv')

# Label fix 
def fix_label(label):
    if label in [0, 1]:
        return 'fake'
    else:
        return 'authentic'

train['label_fixed'] = train['Label'].apply(fix_label)
val['label_fixed'] = val['Label'].apply(fix_label)
test['label_fixed'] = test['Label'].apply(fix_label)

# Merge 
all_df = pd.concat([train, val, test], ignore_index=True)

# Clean 
all_df = all_df.dropna(subset=['Headline', 'Content'])
all_df = all_df.rename(columns={
    'Headline': 'headline',
    'Content': 'content',
    'label_fixed': 'label'
})
all_df = all_df[['headline', 'content', 'label']]

# Save 
all_df.to_csv('C:/Users/Riyad/projects/fake_news/step1_merged.csv', index=False)

# Check 
print("=== Shape ===")
print(all_df.shape)

print("\n=== Labels ===")
print(all_df['label'].value_counts())

print("\n=== Sample ===")
print(all_df.head(3).to_string())
```

```python
import pandas as pd
import re

# Load 
df = pd.read_csv('C:/Users/Riyad/projects/fake_news/step1_merged.csv')

def extract_metadata(df):
    
    # Headline Features 
    df['headline_length'] = df['headline'].str.len()
    df['headline_word_count'] = df['headline'].str.split().str.len()
    df['headline_has_exclamation'] = df['headline'].str.contains('!').astype(int)
    df['headline_has_question'] = df['headline'].str.contains('?', regex=False).astype(int)
    df['headline_has_number'] = df['headline'].str.contains(r'\d').astype(int)
    df['headline_caps_ratio'] = df['headline'].apply(
        lambda x: sum(1 for c in str(x) if c.isupper()) / max(len(str(x)), 1)
    )

    # Content Features 
    df['content_length'] = df['content'].str.len()
    df['content_word_count'] = df['content'].str.split().str.len()
    df['sentence_count'] = df['content'].str.count('[।.!?]')
    df['number_count'] = df['content'].str.count(r'\d+')
    df['punctuation_count'] = df['content'].str.count(r'[।,;:!?"\']')
    df['avg_word_length'] = df['content'].apply(
        lambda x: round(sum(len(w) for w in str(x).split()) / max(len(str(x).split()), 1), 2)
    )
    df['unique_word_ratio'] = df['content'].apply(
        lambda x: round(len(set(str(x).split())) / max(len(str(x).split()), 1), 2)
    )

    # Ratio Features 
    df['content_headline_ratio'] = (
        df['content_word_count'] / df['headline_word_count'].replace(0, 1)
    ).round(2)

    return df

# Apply metadata extraction
df = extract_metadata(df)

# Save 
df.to_csv('C:/Users/Riyad/projects/fake_news/step2_with_metadata.csv', index=False)

# Check 
print("=== Columns ===")
print(df.columns.tolist())

print("\n=== Shape ===")
print(df.shape)

print("\n=== Fake vs Authentic Metadata ===")
meta_cols = [
    'headline_length', 'headline_word_count',
    'content_length', 'content_word_count',
    'unique_word_ratio', 'sentence_count'
]
print(df.groupby('label')[meta_cols].mean().round(2))
```

```python
import pandas as pd
from groq import Groq
import time
from tqdm import tqdm
import os
from dotenv import load_dotenv
from groq import Groq



# API Configuration


API_KEY = os.getenv("GROQ_API_KEY")

client = Groq(api_key=API_KEY)


# Load Dataset

df = pd.read_csv('C:/Users/Riyad/projects/fake_news/step2_with_metadata.csv')

# Extract only fake news for rewriting
fake_df = df[df['label'] == 'fake'].reset_index(drop=True)
print(f"Total fake news found: {len(fake_df)}")


# AI Rewrite Function
# Rewrites each fake news article using LLM
# to create AI-generated fake news samples

def ai_rewrite(headline, content):
    prompt = f"""তুমি একজন বাংলা সাংবাদিক।
নিচের খবরটি পড়ো এবং তথ্য পরিবর্তন করে
বিভ্রান্তিকরভাবে পুনরায় লেখো।
শুধু পুনর্লিখিত খবর দাও, অন্য কিছু না।

শিরোনাম: {headline}
বিষয়বস্তু: {content[:300]}"""

    try:
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",  
            messages=[{"role": "user", "content": prompt}],
            max_tokens=400,
            temperature=0.9
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        print(f"API Error: {e}")
        time.sleep(5)
        return None


# Generate AI Fake News
# Start with 100 samples for testing

HOW_MANY = 100
results = []

for i, row in tqdm(
    fake_df.head(HOW_MANY).iterrows(),
    total=HOW_MANY,
    desc="Generating AI Fake News"
):
    ai_text = ai_rewrite(row['headline'], row['content'])

    if ai_text and len(ai_text) > 30:
        results.append({
            'headline': row['headline'],
            'content': ai_text,
            'label': 'ai_fake'
        })

    # Avoid hitting API rate limits
    time.sleep(0.5)


# Save Generated Dataset

ai_df = pd.DataFrame(results)
ai_df.to_csv(
    'C:/Users/Riyad/projects/fake_news/step3_ai_fake.csv',
    index=False
)

print(f"\nTotal AI fake news generated: {len(ai_df)}")
print(ai_df.head(3).to_string())
```
